#Covid-19 World Vaccination Dataset.
Pre-processing / Cleaning, Analysing and Visualizing the COVID - 19
world vaccination dataset using Python (Pandas, Matplotlib, Seaborn, Plotly)

```python
import pandas as pd
```

```python
#Reading unclean dataset for analysis.
df = pd.read_csv('country_vaccinations.csv')
```

```python
#Checking top few rows to get an idea about the data present in the dataset.
df.head(10)
```

```python
df.info()
```

```python
#Chnging the date column to the correct datatype
df['date'] = pd.to_datetime(df['date'])
```

```python
df.describe()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
df.shape
```

```python
#Check the column names
df.columns
```

```python
#Fill the null values with 0.
# It is very important to know what values to replace the null values with (eg: 0 / mean / median / mode)
'''
As we have cummulative data that gets added day by day, the values on the following day already include the entries from the previous days.
in this case the missing values need to be replaced with 0. Some days don't have any entries as data might be recorded after a fixed interval.
'''
df.fillna(0, inplace=True)
```

```python
#Checking if all null values have been filled.
df.isnull().sum()
```

```python
df.head()
```

```python
#Checking and dropping off the duplicate values.
df.drop_duplicates(inplace=True)
```

```python
#Dropping off the unnecessary columns
df.drop(columns=['source_name', 'source_website'], inplace=True)
```

```python
df.shape
#found no duplicates
```

```python
#Storing the cleaned dataset to make a copy.
clean_df = df.copy()
```

```python
#Storing the cleaned dataset to a CSV file for the later use.
df.to_csv('clean_df.csv')
```

```python
#Creating a dataframe with countries and the vaccines available in those countries.
vax_df = df[['country', 'vaccines']].copy()
```

```python
vax_df.head()
```

```python
vax_df.shape
```

```python
# As we have multiple values in the vaccines column, we need to split them and store them in a list.
vax_df['vaccines'] = vax_df['vaccines'].str.split(', ')
```

```python
vax_df.head()
```

```python
# modify vaccine dataframe to ensure only one entry in each row for the vaccines columns. df.explode() will turn the each vaccine into new row.
# rest of the columns and index values get duplicated
vax_df=vax_df.explode('vaccines')
```

```python
#Notice how index values have been duplicated as well and we need to reset them.
vax_df.head()
```

```python
vax_df.shape
```

```python
#Reset the index values.
vax_df = vax_df.reset_index(drop=True)
```

```python
vax_df.head()
```

```python
vax_df.shape
```

```python
# Checking for the unique values for the vaccine names
vax_df.vaccines.unique()
```

```python
# Checking the number of unique values.
vax_df.vaccines.nunique()
```

```python
#Grouping the dataframe by Vaccine name would give you the list of countries they are available in
grouped_vax = vax_df.groupby('vaccines')['country'].unique().reset_index()
grouped_vax.columns = ['Vaccine', 'Countries Available In']
grouped_vax
```

```python
# Using .nunique() to get a dataframe of vaccines and number of countries they are available in.
grouped_vax_count = vax_df.groupby('vaccines')['country'].nunique().reset_index()
grouped_vax_count.columns = ['Vaccine', 'Countries Available In']
grouped_vax_count
```

# Analyzing the Clean_df.csv to analyse the overall trends and get an overview.

```python
clean_df = pd.read_csv('clean_df.csv')
```

```python
clean_df.head()
```

```python
#Using df.groupby() to get the country - wise data.
#Use Max as aggregate function because we have cummulative data recorded day by day with numbers growing everytime.
#The new values already have the previous values added in them. So, we need the max values to analyse the whole dataset.

summary_df = clean_df.groupby('country').agg({
    'total_vaccinations': 'max',
    'people_vaccinated': 'max',
    'people_fully_vaccinated': 'max',
    'daily_vaccinations': 'mean',
    'daily_vaccinations_per_million': 'mean'
}).reset_index().sort_values(by='total_vaccinations', ascending=False)

summary_df.head()
```

```python
# We need to change the number formatting for the float values as the numbers in above table are harder to read.
pd.options.display.float_format = '{:.0f}'.format
summary_df
```

```python
#Selecting top 10 countries by vaccination numbers and visualizing them with a bar chart.

x = summary_df['country'][:10]
y = summary_df['total_vaccinations'][:10]
plt.figure(figsize=(25, 10))
plt.bar(x, y)
plt.xlabel('Country')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
# Add data labels
for index, row in summary_df.head(10).iterrows():
    plt.text(row['country'], row['total_vaccinations'], row['total_vaccinations'], ha = 'center', va = 'bottom')
plt.show()
```

```python
#Selecting bottom 10 countries according to the vaccination numbers.
bottom_10 = summary_df.tail(10).sort_values(by='total_vaccinations')
bottom_10
```

```python
#Visualizing the 10 coutries with lowest number of vaccination.
x = bottom_10['country']
y = bottom_10['total_vaccinations']
plt.figure(figsize=(25, 10))
plt.bar(x, y)
plt.xlabel('Country')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.title('Bottom 10 Countries with the Lowest Total Vaccinations')

# Add data labels
for index, row in bottom_10.iterrows():
    plt.text(row['country'], row['total_vaccinations'], row['total_vaccinations'], ha = 'center', va = 'bottom')

plt.show()
```
